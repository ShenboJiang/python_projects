```python
#this function implement Column mean imputation to the dataframe
def median_imputation(df):
    imputed_df = df.fillna(df.median())
    return imputed_df
```

```python
#creating a new dataframe
def delete_champion_side(df):
    new_df = df.drop(columns = ["champion", "side"])
    return new_df
```

```python
# this function turn all binary categorical features into 0 and 1
def categorical_to_01(df):
    df = df.replace("Other", 0)
    df = df.replace("TopLane_Jungle", 1)
    df = df.replace("Many", 0)
    df = df.replace("Few", 1)
    return df
```

```python
#domian knowledge feature selection part 2 : delete feature of kill, death and assists, d_spell and f_spell as they do not contribut or have duplicate meaning with other features 
def domain_feature_selection(df):
    new_df = df.drop(columns = ["d_spell", "f_spell", "assists", "deaths", "kills"])
    return new_df
```

```python
#input original df and output after pre-processing df
def pre_processing(df):
    new_df = median_imputation(df)
    new_df = delete_champion_side(new_df)
    pd.options.mode.chained_assignment = None  # default='warn'
    new_df = numerical_to_categorical(new_df, 3, 16)
    new_df = categorical_to_01(new_df)
    new_df = domain_feature_selection(new_df)    
    return new_df
```

```python
#this function apply pi square analysis for feature selection
def feature_selection_xsquare(df_input):
    import pandas as pd
    import numpy as np
    from matplotlib import pyplot as plt
    import scipy.stats as stats
    from scipy.stats import chi2_contingency
    # this is the same table of data from above 
    # set our alpha significance level
    ALPHA = 0.05 #95% confidence interval
    filtered_features = []
    data = df_input.drop(columns = ["level"])
    for feature in data.columns:
        cont_table = pd.crosstab(df_input["level"], df_input[feature])
        chi2_val, p, dof, expected = stats.chi2_contingency(cont_table.values, correction=False)    
        print(f'Chi2 value for feature "{feature}": {chi2_val:.4f}')    
        if(p < ALPHA): 
            filtered_features.append(feature)
            print(f'Reject the Null Hypothesis for feature "{feature}", p-value {p:.4f}\n')
        else:
            print(f'Fail to reject the Null Hypothesis for feature "{feature}", p-value {p:.4f}\n')          
    print('Feature set after filtering with Chi-square test:', filtered_features)
    return filtered_features
```

```python
#this function spliting the input_featute_df and performance_df into training and testing sets
def spliting_training_testing(input_feature_df):
    input_feature_df_training = input_feature_df[0:15000]
    input_feature_df_testing = input_feature_df[15000:len(input_feature_df)]
    return input_feature_df_training, input_feature_df_testing
```

```python
#this function does 5 fold validation with decision tree classifier 
def average_accuracy_with_decisiontree(input_df):
    import numpy as np
    X_origin = input_df[['damage_objectives',
    'damage_building',
    'damage_turrets',
    'kda',
    'time_cc',
    'damage_taken',
    'turret_kills',
    'vision_score',
    'damage_total',
    'gold_earned',
    'role',
    'minions_killed']]
    y_origin = input_df["level"]
    X = X_origin.to_numpy()
    y = y_origin.to_numpy()
    from sklearn.model_selection import KFold
    from sklearn.preprocessing import StandardScaler
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import accuracy_score
    # set k=5 value for k-fold CV
    k = 100
    kf_CV = KFold(n_splits=k, shuffle=True, random_state=42)
    results = []
    for train_idx, test_idx in kf_CV.split(X):
        # train-test split
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        y_train_df = pd.DataFrame(y_train, columns=["level"])
        y_test_df = pd.DataFrame(y_test, columns=["level"])

        X_train_df = pd.DataFrame(X_train, columns=['damage_objectives',
    'damage_building',
    'damage_turrets',
    'kda',
    'time_cc',
    'damage_taken',
    'turret_kills',
    'vision_score',
    'damage_total',
    'gold_earned',
    'role',
    'minions_killed'])
        X_test_df = pd.DataFrame(X_test, columns=['damage_objectives',
    'damage_building',
    'damage_turrets',
    'kda',
    'time_cc',
    'damage_taken',
    'turret_kills',
    'vision_score',
    'damage_total',
    'gold_earned',
    'role',
    'minions_killed'])
        # Training
        model = DecisionTreeClassifier()
        model.fit(X_train_df, y_train_df)
        # Predictions
        y_pred = model.predict(X_test)
        results.append(accuracy_score(y_test, y_pred))
    print(f"Average Accuracy: {np.mean(results)}")
```

```python
#this function does 6 fold validation with k nn classifier 
def average_accuracy_with_knn(input_df, k, n):
    import numpy as np
    X_origin = input_df[['damage_objectives',
    'damage_building',
    'damage_turrets',
    'kda',
    'time_cc',
    'damage_taken',
    'turret_kills',
    'vision_score',
    'damage_total',
    'gold_earned',
    'role',
    'minions_killed']]
    y_origin = input_df['level']
    X = X_origin.to_numpy()
    y = y_origin.to_numpy()
    from sklearn.model_selection import KFold
    from sklearn.preprocessing import StandardScaler
    from sklearn.neighbors import KNeighborsClassifier as KNN
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import accuracy_score
    # set k=5 value for k-fold CV
    kf_CV = KFold(n_splits=k, shuffle=True, random_state=42)
    results = []
    for train_idx, test_idx in kf_CV.split(X):
        # train-test split
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        y_train_df = pd.DataFrame(y_train, columns=["level"])
        y_test_df = pd.DataFrame(y_test, columns=["level"])

        X_train_df = pd.DataFrame(X_train, columns=['damage_objectives',
        'damage_building',
        'damage_turrets',
        'kda',
        'time_cc',
        'damage_taken',
        'turret_kills',
        'vision_score',
        'damage_total',
        'gold_earned',
        'role',
        'minions_killed'])
        X_test_df = pd.DataFrame(X_test, columns=['damage_objectives',
        'damage_building',
        'damage_turrets',
        'kda',
        'time_cc',
        'damage_taken',
        'turret_kills',
        'vision_score',
        'damage_total',
        'gold_earned',
        'role',
        'minions_killed'])
        # Training
        knn = KNN(n_neighbors= n)
        knn.fit(X_train_df, y_train_df)    
        # Predictions
        y_pred = knn.predict(X_test)
        results.append(accuracy_score(y_test, y_pred))
    return(np.mean(results))
```

```python
#this function train a model with k nn classifier and then calcualte the accuracy of the test set
def accuracy_with_knn_actual_model(input_df, testing_set, nbrs):
    import numpy as np
    train_input = input_df.drop(columns = ["level"])
    train_output = input_df['level']
    train_input_np = train_input.to_numpy()
    train_output_np = train_output.to_numpy()
    


    test_input = testing_set[train_input.columns]
    test_output = testing_set['level']
    test_input_np = test_input.to_numpy()
    test_output_np = test_output.to_numpy()

    from sklearn.model_selection import KFold
    from sklearn.preprocessing import StandardScaler
    from sklearn.neighbors import KNeighborsClassifier as KNN
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import accuracy_score
    # Training
    knn = KNN(n_neighbors= nbrs)
    knn.fit(train_input_np, train_output_np)    
    # calculate accuracy through prediction
    output_predict = knn.predict(test_input_np)
    accuracy_score = float(accuracy_score(test_output_np, output_predict))
    # print(f"Accuracy: {accuracy_score}")
    return accuracy_score
```

```python
#this function train a model with k nn classifier and then calcualte the accuracy of the test set
def final_accuracy_with_knn_model(input_df, testing_set, nbrs):
    import numpy as np
    input_df_ = input_df[['damage_building', 'vision_score', 'damage_total', 'gold_earned', 'role', 'minions_killed', "level"]]
    testing_set_ = testing_set[['damage_building', 'vision_score', 'damage_total', 'gold_earned', 'role', 'minions_killed', "level"]]
    train_input = input_df_.drop(columns = ["level"])
    train_output = input_df_['level']
    train_input_np = train_input.to_numpy()
    train_output_np = train_output.to_numpy()
    


    test_input = testing_set_[train_input.columns]
    test_output = testing_set_['level']
    test_input_np = test_input.to_numpy()
    test_output_np = test_output.to_numpy()

    from sklearn.model_selection import KFold
    from sklearn.preprocessing import StandardScaler
    from sklearn.neighbors import KNeighborsClassifier as KNN
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import accuracy_score
    # Training
    knn = KNN(n_neighbors= nbrs)
    knn.fit(train_input_np, train_output_np)    
    # calculate accuracy through prediction
    output_predict = knn.predict(test_input_np)
    accuracy_score = float(accuracy_score(test_output_np, output_predict))
    # print(f"Accuracy: {accuracy_score}")
    return accuracy_score
```

```python
import itertools
def get_combination(feature_list, pack_size):
    #Input: 
    #- features, a list of features from lol data
    #- permutation_size, the size of each permutation
    #Returns a list of lists containing every possible permutation (of size 
    #permutation_size) of candidates
    lst = []
    min_len = min(pack_size, len(feature_list))
    for pack in itertools.combinations(feature_list, min_len):
        lst.append(list(pack))
    return lst
    
def get_combinations_for_features(df):
    input_df = df.drop(columns = ["level"])
    combination_list = []
    for size in range(1,len(input_df.columns)+1):
        list_for_current_size = get_combination(input_df.columns, size)
        for current_combination in list_for_current_size:
            combination_list.append(current_combination)
    return combination_list

combination_of_all_features = get_combinations_for_features(training_set)
highest_accur = 0
for combination_of_feature in combination_of_all_features:
    current_trainning_set = training_set[combination_of_feature + ["level"]]
    current_accuracy_score = accuracy_with_knn_actual_model(current_trainning_set, testing_set, 24)

    if highest_accur < current_accuracy_score:
        highest_accur = current_accuracy_score
        highest_score_combination = combination_of_feature
        print(highest_accur)
print("highest combination accuracy is " + str(highest_accur) + "the highest feature input combination is " + str(highest_score_combination))
```

```python
def plot_scatter_bestfit(x, y):
    
    import matplotlib.pyplot as plt
    from scipy import stats
    new_x = median_imputation(x)
    new_y = median_imputation(y)
    slope, intercept, r, p, std_err = stats.linregress(new_x, new_y)
    def myfunc(new_x):
        return slope * new_x + intercept
    mymodel = list(map(myfunc, new_x))
    plt.scatter(new_x, new_y, color='dodgerblue', s=10)
    plt.plot(new_x, mymodel, color='tomato')
    plt.ylim((0, 25))
    return plt

plt = plot_scatter_bestfit(raw_df_after_median_inputation["gold_earned"], raw_df_after_median_inputation["level"])
#plt.savefig('books_read.png', dip = 1200)
plt.show()
```
